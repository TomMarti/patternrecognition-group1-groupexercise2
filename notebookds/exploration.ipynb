```python
import random
import matplotlib.pyplot as plt

from src.preprocess import build_word_image_index

DATA_ROOT = PROJECT_ROOT / "data"

# Schritt 1 ausf체hren
train_images = build_word_image_index(DATA_ROOT, "train")

print("Total train images:", len(train_images))

# 5 zuf채llige Word-IDs anzeigen
sample_ids = random.sample(list(train_images.keys()), 5)

plt.figure(figsize=(12, 4))

for i, wid in enumerate(sample_ids):
    img = train_images[wid]

    plt.subplot(1, 5, i + 1)
    plt.imshow(img, cmap="gray")
    plt.axis("off")
    plt.title(wid)

plt.show()
```

```python
def show_word(word_id, word_images):
    import matplotlib.pyplot as plt
    import numpy as np

    if word_id not in word_images:
        print(f"Word ID '{word_id}' not found!")
        return

    img = word_images[word_id]

    plt.figure(figsize=(5, 3))
    plt.imshow(img, cmap="gray")
    plt.title(word_id)
    plt.axis("off")
    plt.show()

show_word("270-18-01", train_images)
show_word("277-25-06", train_images)
show_word("274-34-06", train_images)
show_word("279-14-04", train_images)
show_word("277-11-02", train_images)
```
